```python
import time

class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password

class Quiz:
    def __init__(self, title, description, duration):
        self.title = title
        self.description = description
        self.duration = duration
        self.questions = []
    
    def add_question(self, question, options, correct_answer):
        question_data = {
            'question': question,
            'options': options,
            'correct_answer': correct_answer
        }
        self.questions.append(question_data)

    def display_quiz(self):
        print(f"Quiz Title: {self.title}")
        print(f"Description: {self.description}")
        print(f"Duration: {self.duration} minutes\n")
        
        for idx, question_data in enumerate(self.questions, start=1):
            print(f"Question {idx}: {question_data['question']}")
            for option_idx, option in enumerate(question_data['options'], start=1):
                print(f"{option_idx}. {option}")
            print()

    def take_quiz(self):
        score = 0
        start_time = time.time()

        for question_data in self.questions:
            print(question_data['question'])
            for idx, option in enumerate(question_data['options'], start=1):
                print(f"{idx}. {option}")
            
            user_answer = int(input("Your answer (Enter the option number): "))
            correct_answer = question_data['correct_answer']

            if user_answer == correct_answer:
                score += 1
        
        end_time = time.time()
        time_taken = end_time - start_time
        print("\nQuiz completed!")
        print(f"Your score: {score}/{len(self.questions)}")
        print(f"Time taken: {time_taken:.2f} seconds")

if __name__ == "__main__":
    users = []  # List to store user objects
    logged_in_user = None

    def register():
        print("User Registration:")
        username = input("Enter a username: ")
        password = input("Enter a password: ")
        user = User(username, password)
        users.append(user)
        print("Registration successful!\n")

    def login():
        print("User Login:")
        username = input("Enter your username: ")
        password = input("Enter your password: ")

        for user in users:
            if user.username == username and user.password == password:
                print(f"Welcome, {username}!\n")
                return user
        
        print("Invalid username or password. Please try again.\n")
        return None

    def create_quiz():
        print("Quiz Creation:")
        title = input("Enter quiz title: ")
        description = input("Enter quiz description: ")
        duration = int(input("Enter quiz duration in minutes: "))

        quiz = Quiz(title, description, duration)
        num_questions = int(input("Enter the number of questions for the quiz: "))

        for i in range(num_questions):
            print(f"Question {i+1}:")
            question = input("Enter the question: ")
            options = []
            for j in range(4):
                options.append(input(f"Enter option {j+1}: "))
            correct_answer = int(input("Enter the correct option number: "))
            quiz.add_question(question, options, correct_answer)
        
        print("Quiz created successfully!\n")
        return quiz

    def take_quiz():
        if not logged_in_user:
            print("Please login first to take a quiz.\n")
            return

        #example:
        quiz = create_quiz()  
        quiz.display_quiz()
        quiz.take_quiz()

    def view_results():
        if not logged_in_user:
            print("Please login first to view your results.\n")
            return

        # Code to display the user's past quiz results
        # You can implement this part based on your preferred user interface

        # For example:
        print("No past quiz results available.\n")

    while True:
        print("Quiz Application:")
        print("1. Register")
        print("2. Login")
        print("3. Create Quiz")
        print("4. Take Quiz")
        print("5. View Results")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            register()
        elif choice == "2":
            logged_in_user = login()
        elif choice == "3":
            if not logged_in_user:
                print("Please login first to create a quiz.\n")
            else:
                create_quiz()
        elif choice == "4":
            take_quiz()
        elif choice == "5":
            view_results()
        elif choice == "6":
            print("Exiting the Quiz Application. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
```
