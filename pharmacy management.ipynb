```python
users = {
    "dholu@gmail.com": "password1",
    "bholu@gmail.com": "password2",
    # Add more users here
}

#Medicines to choose from
medicines = {
    1: {"name": "Medicine 1", "price": 100},
    2: {"name": "Medicine 2", "price": 200},
    
}

#Example
user_cart = {}

def login():
    email = input("Enter your email: ")
    password = input("Enter your password: ")
    if email in users and users[email] == password:
        return email
    else:
        print("Invalid email or password.")
        return None

def view_medicine_list():
    print("Available Medicines:")
    for med_id, med_info in medicines.items():
        print(f"{med_id}. {med_info['name']} - ${med_info['price']}")

def add_to_cart(user_email, medicine_id):
    if user_email not in user_cart:
        user_cart[user_email] = []
    user_cart[user_email].append(medicine_id)

def view_cart(user_email):
    return user_cart.get(user_email, [])

def calculate_total_price(user_email):
    cart_items = view_cart(user_email)
    total_price = 0
    for item_id in cart_items:
        total_price += medicines[item_id]["price"]
    return total_price

def main():
    user_email = login()
    if user_email:
        while True:
            print("\n1. View Medicines")
            print("2. Add Medicine to Cart")
            print("3. View Cart")
            print("4. Checkout")
            print("5. Logout")

            choice = input("Enter your choice: ")

            if choice == "1":
                view_medicine_list()
            elif choice == "2":
                medicine_id = int(input("Enter the ID of the medicine you want to add to the cart: "))
                if medicine_id in medicines:
                    add_to_cart(user_email, medicine_id)
                else:
                    print("Invalid medicine ID.")
            elif choice == "3":
                cart_items = view_cart(user_email)
                print("Cart items:", cart_items)
            elif choice == "4":
                total_price = calculate_total_price(user_email)
                print("Total Price:", total_price)
            elif choice == "5":
                break
            else:
                print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
